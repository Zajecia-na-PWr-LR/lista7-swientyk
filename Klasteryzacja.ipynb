### Import danych

```python
from scipy.io import arff
import pandas as pd
import numpy as np

source = arff.loadarff('/home/jakub/ml/lista7-swientyk/Dry_Bean_Dataset.arff')
                       
df = pd.DataFrame(source[0])
```

```python
print("liczba duplikatów:")
print(df.duplicated().sum())
df[df.duplicated(keep=False)]
```

```python
df = df.drop_duplicates()
```

```python
print("liczba duplikatów po usunięciu:")
print(df.duplicated().sum())
```

```python
df['Class'] = df['Class'].str.decode('utf-8') 
```

### Macierz korelacji

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
corr_data = df.select_dtypes(['float64', 'int64']).corr()

plt.figure(figsize = (15,10))
sns.heatmap(data=corr_data, 
            vmin=-1,
            vmax=1,
            cmap='coolwarm',
            annot=True,
            fmt=".2f")

plt.title('Macierz korelacji')
```

### Redukcja wymiarów
wyrzucam cechy skorelowane z innymi >= 85%

```python
drop_area = ['Perimeter', 'MajorAxisLength', 'MinorAxisLength', 'ConvexArea', 'EquivDiameter', 'ShapeFactor1']
df = df.drop(columns=drop_area)
drop_aspectratio = ['Eccentricity', 'Compactness', 'ShapeFactor3']
df = df.drop(columns=drop_aspectratio)
```

```python
corr_data = df.select_dtypes(['float64', 'int64']).corr()

plt.figure(figsize = (10,6))
sns.heatmap(data=corr_data, 
            vmin=-1,
            vmax=1,
            cmap='coolwarm',
            annot=True,
            fmt=".2f")

plt.title('Macierz korelacji')
```

### Rozkład klas

```python
sns.countplot(df, x=df.Class, hue='Class')
plt.title('Rozkład klas')
plt.show()
```

```python
sns.boxplot(data=df, x='Class', y='Area')
```

### Wizualizacje dystrybucji poszczególnych cech

```python
sns.scatterplot(data=df, hue='Class', x='roundness', y='Area', alpha=0.4, edgecolor=None, s=17)
plt.grid(ls='--')
plt.title('Dystrybucja - powierzchnia i okrąłość')
```

```python
sns.scatterplot(data=df, hue='Class', x='ShapeFactor2', y='ShapeFactor4', alpha=0.4, edgecolor=None, s=17)
plt.grid(ls='--')
plt.title('Dystrybucja - dwa pozostałe czynniki kształtu')
```

```python
sns.scatterplot(data=df, hue='Class', x='Solidity', y='Extent', alpha=0.4, edgecolor=None, s=17)
plt.grid(ls='--')
plt.title('Dystrybucja - Zwartość i zasięg (słabo skorelowane)')
```

```python
sns.scatterplot(data=df, hue='Class', x='AspectRation', y='ShapeFactor2', alpha=0.4, edgecolor=None, s=17)
plt.grid(ls='--')
plt.title('Dystrybucja - Współczynnik proporcji i jakiś wyznacznik kształtu (silnie skorelowane)')
```
